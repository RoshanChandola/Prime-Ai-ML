```python
# data concatenation is used to combine data when we have data in multiple parts
import pandas as pd
```

```python
df1=pd.DataFrame({
    "id":[1,2,3,4],
    "name":["roshan","uday","amit", "vedaant"]
})

df2=pd.DataFrame({
    "id":[5,6,7,8],
    "name":["adit","naman","aman", "suman"]
    })
pd.concat([df1,df2])
pd.concat([df1,df2], ignore_index=True)
pd.concat([df1,df2],axis=1,ignore_index=True
         )
```
