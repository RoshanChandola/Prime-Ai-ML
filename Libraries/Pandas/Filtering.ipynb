```python
import pandas as pd
# importing data with some specific requirements or following condition
```

```python
df= pd.read_csv("globalAirQuality.csv")

df[df["aqi"]>100]# filtering values based on condition

#combining multiple condition 
df[(df["aqi"]>100)& (df["temperature"]>30) ]
# df[(df["aqi"]>100)& (df["temperature"]<30) ]

# chaining

aqi_data=df[( df["aqi"] > 100) & (df["temperature"]>30)] [["city","aqi"]]# getting only the columns we want

aqi_data
aqi_data.iloc[0]
aqi_data.loc[6]
```

```python
#query method to filtering data
df.query("aqi>100 & temperature >30")[["city","aqi"]]
```

```python
aqi_value=100
aqi_data=df.query("aqi>@aqi_value").copy# using variables
print(aqi_data) 
#query returns copy of data
```
