```python
import pandas as pd
df=pd.DataFrame(
    {
        "country": ["India","USA","China","USA"],
        "year": [2020,2021,2020,2021],
        "sales": [100,120,90,110],
        "profit": [20,34,65,32],
     
    })
melted_data=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrices",
    value_name="value")
melted_data

# #pivot function
melted_data.pivot_table(
    index=["country", "year"],
    columns="metrices",
    values="value",
)
```
