```python
""" we have different types of functions to perform feature engineering
in which we create some new columns to increase performance of model 
apply(): writing logic in a function then passing as refrence 
map():changing in existing data
assign(): creating new columns with assigning them new value
replace(old,new): replace old value with new """
import pandas as pd
```

```python
#feature Engineering
#high order function in which we pass function as arguments

#using apply funciton 
df=pd.read_csv("raw_data.csv")

df2=df.copy()
df2["tax"] = df2["income"].apply(lambda x:"20%" if x>=60000 else "10%")

#map(): changing data using dictionary
gender_map={"Male": "M", "Female":"F", "Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map)

#assign function
df2=df2.assign(new_income=df2["income"]*1.1)

#replace(old,new)
df2["country"]=df2["country"].replace("USA","United States")
```
