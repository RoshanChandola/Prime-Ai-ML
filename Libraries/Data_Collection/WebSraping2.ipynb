```python
import requests 
from bs4 import BeautifulSoup
```

```python
#fetch data and store it
page_count = 1

while True:
    try:
        url = f"https://quotes.toscrape.com/page/{page_count}/"
        res = requests.get(url, timeout=10)
        res.raise_for_status()   # raises HTTPError for 4xx/5xx

        soup = BeautifulSoup(res.text, "lxml")
        quotes = soup.select("div.quote")

        if not quotes:
            print("No valid pages anymore")
            break

        with open(
            f"E:\\Prime_Ai_ML\\AdvanceLearning\\Data_Collection\\scraped_data\\quotes{page_count}.html",
            "w",
            encoding="utf-8"
        ) as f:
            f.write(res.text)

        print(f"Downloaded data from page {page_count}")
        page_count += 1

    # üîπ Network-related errors
    except requests.exceptions.RequestException as e:
        print(f"Request failed on page {page_count}: {e}")
        break

    # üîπ File-related errors
    except FileNotFoundError:
        print("File path not found. Check directory structure.")
        break

    # üîπ Any unexpected error
    except Exception as e:
        print(f"Unexpected error: {e}")
        break
```

```python
#extract useful info - pages1

with open("E:\\Prime_Ai_ML\\AdvanceLearning\\Data_Collection\\scraped_data\\quotes1.html","r",  encoding="utf-8") as f:
    html_content= f.read()
    
soup= BeautifulSoup(html_content, "lxml")
```

```python
all_quotes = soup.select("div.quote")
life_quotes = []

for q in all_quotes:
    all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text, author])

print(life_quotes)
```
