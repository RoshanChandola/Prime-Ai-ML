```python
#numpy is a open source python library stands for numerical python basically
#basically faster than python list and contains homogeneous type of data in contigous memory location 
#optimized in c language(because c is a low level language (more closer to system))
import numpy as np 
import time
```

```python
#creating a numpy array 
arr=np.array([1,2,4,4])
print(arr,type(arr))
```

```python
#comparing performance of numpy array and list 
# execution performance
size=1_000_000
# python list
py_list=list(range(size))
start=time.time()
sq_list=[x**2 for x in py_list]
end=time.time()
print(f"python list time={end-start} seconds")
#numpy arrays
np_arr=np.array(py_list)
start=time.time()
sq_array=np_arr**2
print(f"python numpy array time={end-start} seconds")
    
```

```python
#comparing in terms of memory
# numpy arrays uses less memory then list becuase python list stores more metadata 
import sys
print(f"python list size={sys.getsizeof (py_list)*len(py_list)}bytes")
print(f"python numpy array size={np_arr.nbytes}bytes")

```
