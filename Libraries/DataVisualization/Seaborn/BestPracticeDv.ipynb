```python
import seaborn as sns
import matplotlib.pyplot as plt

tips=sns.load_dataset("tips")
fig,ax=plt.subplots()
sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    # size="total_bill",
    ax=ax,
    ci=None,errorbar=None
)
ax.set_title("Avg bill value for each day")
ax.set_xlabel("days")
ax.set_ylabel("total bill amount")
fig.tight_layout()
tips.head()
```
