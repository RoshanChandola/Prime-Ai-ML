```python
import json
```

```python
#Load the data
def load_data(filename):
    with open (filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("Store_data.json")
print(data)
print(type(data))
```

```python
#clean &  structure the data
def clean_data(data):
    text_to_num={"one":1, "two":2,"three":3 ,"four":4, "five":5} 
    cleaned_data=[]
    unique_users=set()
    for user in data:
       raw_rating=str(user["rating"]).strip().lower()
       if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
       user["rating"]=raw_rating
        #handel the missing vals
       raw_age=user.get("age")
       if(raw_age==None):
            user["age"]=None
        #Deduplication
       if(user["name"].strip()  in unique_users):
            continue
       unique_users.add(user["name"])
       cleaned_data.append(user)
    return cleaned_data
        
```

```python
data=clean_data(data)
```

```python
# Get meaningul insights from data
def get_insights(data):
    #avg rating
    total_rating=0
    for  user in data:
        total_rating+= float(user["rating"])
    print(f"avg rating ={total_rating/len(data)}")
    # percentage of users with poor rating
    poor_ratings=0
    for user in data:
        if(float(user["rating"])<3):
            poor_ratings+=1
    print(f"percentage  of user with poor rating ={poor_ratings/len(data)*100}%")


        
```

```python
get_insights(data)
```

```python
# build Recommendation features
def get_recommendations(data):
    recommendations = []

    for users in data:
        curr_recomm = {}
        curr_recomm["name"] = users["name"]

        if float(users["rating"]) >= 4:
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"

        recommendations.append(curr_recomm)

    return recommendations
```

```python
get_recommendations(data)
```
