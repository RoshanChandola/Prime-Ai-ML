```python
# web  Scrapping

import requests as req

```

```python
url="https://www.scrapethissite.com/pages/simple/"
res=req.get(url)
```

```python
if res.status_code==200:
    print(res.headers)

```

```python
#downloading and storing files 
with open("scraped_data/data1.html", "w")as f:
    f.write(res.text)
```
