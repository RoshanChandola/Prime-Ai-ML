```python
import pandas as pd
#a data frame in pandaas is a 2d labled array 
#it is used to show data in tabular form 
```

```python
#creating a dataframe using dictionary
info={
     "NAME": ["roshan","uday","aditi","amit"],
    "CGPA": [9.53,9.37,9.33,9.62]
}
df=pd.DataFrame(info)
print(df)
```

```python
#creating dataframe using nested list
df=pd.DataFrame([["roshan",20],["Vedaant",22]],columns=["name","aGe"])
print(df)
```

```python
#creating using numpy array
import numpy as np
arr=np.array([[1,2,3],[1,2,3]])
df=pd.DataFrame(arr,columns=['A','B','C'])
print(df)
```

```python
"""
Learning about the methods of dataframe
using these we can perform eda on data sets"""
#head method
df.head() #it shows first 5 rows of dataframe
df.head(2) #it shows first 2 rows of dataframe  
#tail method
df.tail() #it shows last 5 rows of dataframe
df.tail(2) #it shows last 2 rows of dataframe
#info method
df.info() #it shows the summary of dataframe
#describe method
df.describe() #it shows the statistical summary of dataframe
#shape method
df.shape #it shows the number of rows and columns in dataframe
#dtypes method
df.dtypes #it shows the data types of each column in dataframe
#columns method
df.columns #it shows the column names of dataframe
#index method
df.index #it shows the index of dataframe
#sort_values method
df.sort_values(by='CGPA') #it sorts the dataframe based on the column name provided
#nunique method
df.nunique() #it shows the number of unique values in each column of dataframe
```
