```python
# plotting real data
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005, 170, 427, 133, 232] # in $M

#plot
plt.plot(years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Linocoln"]
non_oscar_revenue=[378,2788,829,185,275] # in $M
plt.plot(years, oscar_revenue,"o--r", label="Oscar_Movies")
plt.plot(years, non_oscar_revenue,"<-y", label="Non_Oscar_Movies")
plt.title("Oscar Movies Revenue vs Non_oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
```

```python
#changing appreance of graph using format strings and other properties
# Non-Oscar Winner Movies
non_oscar_movies=["Slumdog Millionaire","Avatar","Inception","Hugo","Linocoln"]
non_oscar_revenue=[378,2788,829,185,275] # in $M
plt.plot(years, oscar_revenue,"o--b",linewidth=5, label="Oscar_Movies")# used marker(circle),line(dashed),color(red)
plt.plot(years, non_oscar_revenue,linewidth=5,color="r",marker=">",linestyle=":", label="Non_Oscar_Movies")
plt.title("Oscar Movies Revenue vs Non_oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
```

```python
plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar Movies")
plt.title("Oscar Movies Revenue vs Non_oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.grid()
plt.legend()
print(plt.style.available)
# checks the available styles
plt.style.use("dark_background")
plt.show()
```

```python
# using the function tight_layout()
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue=[1005, 170, 427, 133, 232] # in $M
plt.style.use("default")
plt.plot(oscar_movies, oscar_revenue)
plt.xlabel("movies list")
plt.ylabel("revenue in $m")
plt.tight_layout()
plt.savefig("final_graph.png")
```
