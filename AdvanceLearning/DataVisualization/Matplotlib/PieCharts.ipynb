```python
import matplotlib.pyplot as plt
# pie  Charts
expenses=["Salaries", "Rent", "Marketing", "R&D","Miscellaneous"]
amounts=[500,150,120,100,50]
plt.style.use("default")
explode=[0,0,0,0,0]
plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
       wedgeprops={
       "edgecolor":"black",
       "linewidth":1,
        "linestyle":"--"
       },
        explode=explode,
        startangle=180,
        shadow=True
)

# plt.style.available
```
